```python
import requests
import json

# Endpoint URL (replace with your actual server URL if different)
url = "http://65.108.32.175:8060/video"

# Data to be sent in the request (example values)
data = {
    # "seed": 42,  # Example seed for randomness
    "prompt": "A serene landscape with mountains and a lake"  # Example text prompt
}

# Make the POST request to the server
response = requests.post(url, json=data)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response content
    content = response.json()
    video_base64_string = content['completion']
    print("Video generation successful. Video data (base64):", video_base64_string[:100] + "...")  # Print the first 100 characters to check
else:
    print(f"Failed to generate video. Status code: {response.status_code}, Detail: {response.text}")

```

```python
import base64


decoded_data = base64.b64decode(data)
file_path = "paris.mp4"
with open(file_path, "wb") as file:
    file.write(decoded_data)

print(f"Video saved to {file_path}")
```
